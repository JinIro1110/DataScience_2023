```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('seaborn')
sns.set(font_scale=2.5)

import missingno as msno

import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
```

# 2.1 Pclass

```python
df_train[['Pclass', 'Survived']].groupby(['Pclass'], as_index=True).count()
```

```python
df_train[['Pclass', 'Survived']].groupby(['Pclass']).count()
```

```python
df_train[['Pclass', 'Survived']].groupby(['Pclass']).sum()
```

```python
pd.crosstab(df_train['Pclass'], df_train['Survived'], margins=True).style.background_gradient(cmap='cool')
```

```python
df_train[['Pclass', 'Survived']].groupby(['Pclass'], as_index=True).mean().sort_values(by='Survived', ascending=False).plot.bar(figsize=(18,8))
```

```python
y_position = 1.02
f, ax = plt.subplots(1, 2, figsize=(18, 8))
df_train['Pclass'].value_counts().plot.bar(color=['#CD7F32', '#FFDF00', '#D30303'], ax=ax[0])
ax[0].set_title('Number of passengers By Pclass', y=y_position)
ax[0].set_ylabel('Count')
sns.countplot(x='Pclass', hue='Survived', data=df_train, ax=ax[1])
ax[1].set_title('Pclass: Survived vs Dead', y=y_position)
plt.show()
```

# 2.2 Sex

```python
f, ax = plt.subplots(1, 2, figsize=(18,8))
df_train[['Sex', 'Survived']].groupby(['Sex'], as_index=True).mean().plot.bar(ax=ax[0])
ax[0].set_title('Survived vs Sex')
sns.countplot(data=df_train,x='Sex', hue='Survived', ax=ax[1])
ax[1].set_title('Survived vs Dead')
plt.show()
```

```python
df_train[['Sex', 'Survived']].groupby(['Sex'], as_index=True).mean().plot.bar(figsize=(18,8))
```

```python
pd.crosstab(df_train['Sex'], df_train['Survived'], margins=True).style.background_gradient(cmap='cool')
```
